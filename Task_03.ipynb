# Self-Dividing Number

練習：Self-Dividing Number（自除數）指的是可以被每位數的數字整除的數

    例如，128 可以同時被 1、2、8 分別整除即符合。
本題請根據輸入的上下邊界，找出該區間內的Self-Dividing Number中哪兩個數之間的差距最大。

    舉個例子，假如從11-20符合的數有：11、12、15，
    其中11跟12的差距等於1、12跟15的差距等於3，
    因此差距最大是12跟15。

Requirements：

1. 讓使用者輸入兩個數字，代表一個區間

2. 請利用程式找出該區間中符合 Self-Dividing Number 的所有數

3. 將差距最大的 Self-Dividing Number 相鄰兩數印出

````python
#method 1
a,b = eval(input('please input 2 integer numbers: '))
L = [i for i in range(a,b+1)] #a,b之間的整數
A = []

#找符合的被除數
for i in L:
  n = len(str(i)) #該數幾位數
  Dflag = True
  for j in range(n):
    digits = int(str(i)[j]) #第幾位數的除數
    if digits == 0 or i % digits != 0:  #除數為0或不整除的情況下break
      Dflag = False #可否被整除
      break
  if Dflag == 1:
    A.append(i)
# print(A)  #符合的數值


#找相鄰數值對簡最大值
maxdiff = 0
for i,j in enumerate(A[:-1]): #只找到倒數第2
  k = A[i+1]         #j的下一個
  diff = k-j
  if diff > maxdiff:
    maxdiff = diff
    d = (j,k)
print(d)       
````

````python
#method 2
a,b = eval(input('please input 2 integer numbers: ')) #10,9000
L = list(range(a,b+1)) #a,b之間的整數
A = []

#找符合的被除數
for i in L:
  digits = [int(d) for d in str(i)]             #int(d)該數字轉整數、str(i)產生該數的字串
  if all([a != 0 and i % a == 0 for a in digits]):  #all(list、tuple、dict、etc)判斷內容物的是否，0、false、NaN都為False
    A.append(i)
#print(A)

#找相鄰數值對簡最大值           
diff = [k-j for j,k in zip(A,A[1:])]    #這些自除數兩相鄰數相檢的值，zip(a,b)會找最短內容物的配對
diff_index = diff.index(max(diff))       #找最大差值的index，等等要回推兩數字在A的位置
d = (A[diff_index],A[diff_index+1])
print(d)  #(5775, 6126)
````

# Two Sum

Requirements：

1. 預設題目會提供兩個變數 nums 列表和 target 整數

2. 請找出 nums 中兩數和為 target 值的 index 並存放在列表後印出

````python
nums = [2,7,11,15]
target = 9

N = [3,2,4]
T = 6

N2 = [4,5]
T2 = 10

def twosum(nums,target):
  L = ()  #找過的數
  for i,j in enumerate(nums):
    n = target - j
    if n in L:
      return [nums.index(n),i]
    L += (j,)
d = twosum(nums,target) #[0, 1]
d = twosum(N,T) #[1, 2]
d = twosum(N2,T2) #None
print(d)
````
