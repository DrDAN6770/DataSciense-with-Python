````python
#method1
s = 'Here are UPPERCASE and lowercase chars.'

def method1(s):
  Litter = {}
  for v,k in enumerate(s):
    if k in Litter:
      Litter[k] = Litter[k] + [v+1]
    else:
      Litter.setdefault(k,[v+1])      #dict.setdefault(a,b) a: key值 返回value值；若無a則新增b值，並加入到原dict
  return Litter

print(method1(s))
````

````python
#method2
S = 'Here are UPPERCASE and lowercase chars.'

d = {}
for i, s in enumerate(S):
  d.setdefault(s, [])
  d[s].append(i + 1)

print(d)
````

````python
def add1(n):
  return n+1

def isPrime(n):
  return all(n % i for i in range(2, int(n**.5)+1))

def f(L,F):
  return [F(i) for i in L]
'''
上面拆開就是下面這樣
def f(L,F):
  ans = []
  for i in L:
    ans.append(F(i))
  return ans
'''
# 函式F也可以當成變數或是參數傳入另外一個函式，例如F = isPrime

L = [1,2,3,4,5,6]
F = add1
print(f(L, F)) # [2,3,4,5,6,7]

L = [2,3,4,5,6,7] 
F = isPrime
print(f(L, F)) # [True,True,False,True,False,True]
````
