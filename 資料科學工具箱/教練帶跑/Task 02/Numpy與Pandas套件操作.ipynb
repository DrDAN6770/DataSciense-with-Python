Pandas 是一個 Python 的資料分析套件，提供了高效的資料結構，方便快速地進行資料處理和分析。

Pandas 套件主要有幾種資料結構：Series 和 DataFrame、Panel（少用）。其中，Series 可以視為**一維的陣列**，每個元素有一個對應的索引；DataFrame 則是一個**二維的表格**，每列和每欄都有一個對應的索引。

* Series：一維陣列，類似於帶有標籤的陣列。
* DataFrame：二維表格，類似於 Excel 表格或 SQL 資料表。
* Panel：三維數據，不常用。

#建立Pandas物件與外部資料存取

##Series 物件

```python
import pandas as pd

# 創建一個Series物件
s = pd.Series([1, 3, 5, 7, 9])

# Series物件的索引是從 0 開始的整數，可以通過索引來訪問Series物件中的元素。
print(s)
```

##DataFrame物件

```python
# 創建一個DataFrame物件
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
        'age': [25, 32, 18, 47],
        'city': ['New York', 'Paris', 'London', 'Berlin']}

df = pd.DataFrame(data)

# DataFrame物件的列可以通過列名訪問，行可以通過索引或標籤訪問。
print(df)
```

#Series、DataFrame 的屬性

##Series物件的屬性

---

Series物件有一些常用的屬性，包括values、index、dtype和shape等。

* values屬性返回Series物件中的元素
* index屬性返回索引
* dtype屬性返回數據類型
* shape屬性返回數據形狀

```python
# 創建一個Series物件
s = pd.Series([1, 3, 5, 7, 9])

# 訪問Series物件的屬性
print(s.values)   # [1 3 5 7 9]
print(s.index)    # RangeIndex(start=0, stop=5, step=1)
print(s.dtype)    # int64
print(s.shape)    # (5,)
```

##DataFrame物件的屬性

---

DataFrame物件有一些常用的屬性，包括values、index、columns、dtype和shape等。
* values屬性返回DataFrame物件中的元素
* index屬性返回行索引
* columns屬性返回列索引
* dtype屬性返回數據類型
* shape屬性返回數據形狀

```python
# 創建一個DataFrame物件
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
     'age': [25, 32, 18, 47],
     'city': ['New York', 'Paris', 'London', 'Berlin']}

df = pd.DataFrame(data)

# 訪問DataFrame物件的屬性
print(df.values)
```

```python
print(df.index) 
```

```python
print(df.columns)
```

```python
print(df.dtypes)
```

```python
print(df.shape)
```

###DataFrame 的資料選取

---

DataFrame物件支持多種資料選取方式，包括列選取、行選取和切片選取等。其中列選取和行選取是最常用的資料選取方式

####第一種方式「行選取」可以通過欄位進行選取

```python
# 創建一個DataFrame物件
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
     'age': [25, 32, 18, 47],
     'city': ['New York', 'Paris', 'London', 'Berlin']}

df = pd.DataFrame(data)

# 通過列名選取列
print(df['name'])
```

```python
# 也可以利用切片進行多行選取
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
        'age': [25, 32, 18, 47],
        'city': ['New York', 'Paris', 'London', 'Berlin']}

df = pd.DataFrame(data)

# 通過列名選取多列
print(df[['name', 'city']])
```

####第二種方式「列選取」可以通過索引名稱或索引進行選取

```python
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
     'age': [25, 32, 18, 47],
     'city': ['New York', 'Paris', 'London', 'Berlin']}

df = pd.DataFrame(data)

# 通過行索引選取
print(df[0:3])
```

```python
# 或是利用索引名稱進行選取

df.set_index('name', inplace=True)
print(df['Alice':'Charlie'])
```

####DataFrame 支援一種用座標方式的二維選取

---

以上兩種方法一次選取都只能考慮一個維度

.loc」是一個用於基於標籤（label）選擇行和列的方法，其中 .loc 用於純粹的標籤索引即行和列的名稱，基本語法如下

    df.loc[row_indexer, column_indexer]



*   row_indexer : 選擇行的標籤
*   column_indexer : 選擇列的標籤



---

這兩個參數都可以是一個標籤（單一標籤），也可以是一個標籤列表（多個標籤）。

當 row_indexer 和 column_indexer 都是標籤列表時，返回的是一個子 DataFrame。假設有以下的 DataFrame

```python
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
     'age': [25, 32, 18, 47],
     'city': ['New York', 'Paris', 'London', 'Berlin']}
df = pd.DataFrame(data)
df.set_index('name', inplace=True)

# 選取 "Alice" 和 "Charlie" 兩行的 "age" 和 "city" 這兩列
result = df.loc[['Alice', 'Charlie'], ['age', 'city']]
print(result)
```

####遮罩選取

---

遮罩選取指的是根據一個或多個條件建立一個布林遮罩，然後用這個遮罩選取DataFrame中的數據

```python
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
     'age': [25, 32, 18, 47],
     'city': ['New York', 'Paris', 'London', 'Berlin']}
df = pd.DataFrame(data)
df
```

```python
# 使用遮罩選取
mask = df['age'] > 30  # 創建一個年齡大於30的遮罩
print(df[mask])
```

```python
mask = (df['age'] > 30) & (df['city'] == 'Paris')  # 創建一個年齡大於30且城市為Paris的遮罩
print(df[mask])
```

###DataFrame 的算術運算與排序

---

DataFrame物件支持多種算術運算，包括加法、減法、乘法和除法等

```python
# 創建兩個DataFrame物件
data1 = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
      'age': [25, 32, 18, 47],
      'city': ['New York', 'Paris', 'London', 'Berlin']}
df1 = pd.DataFrame(data1)

data2 = {'name': ['Eva', 'Frank', 'Grace', 'Henry'],
      'age': [29, 23, 37, 31],
      'city': ['Rome', 'Sydney', 'Tokyo', 'Moscow']}
df2 = pd.DataFrame(data2)
```

```python
# 進行加法
print(df1 + df2)

#當DataFrame物件進行算術運算時，如果兩個物件相對應的位置都存在值，則進行運算；否則，該位置的值為NaN
```

```python
# DataFrame物件可以按照指定的列或行進行排序
# 創建一個DataFrame物件
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
     'age': [25, 32, 18, 47],
     'city': ['New York', 'Paris', 'London', 'Berlin']}

df = pd.DataFrame(data)
# 按照age列進行排序，預設為升序排序
print(df.sort_values(by='age'))
print('--------------------------')
# 按照age列進行降序排序
print(df.sort_values(by='age', ascending=False))
print('--------------------------')
# 按照索引進行排序 >> 預設
print(df.sort_index())
```

###DataFrame 資料合併與重組

---

DataFrame物件可以根據指定的條件將不同的DataFrame物件合併在一起



* concat()
* merge()
* join()



####concat()

---
使用 concat() 方法可以將多個DataFrame物件沿著指定的軸合併在一起


```python
# 創建兩個DataFrame物件
data1 = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
      'age': [25, 32, 18, 47],
      'city': ['New York', 'Paris', 'London', 'Berlin']}
df1 = pd.DataFrame(data1)

data2 = {'name': ['Eva', 'Frank', 'Grace', 'Henry'],
      'age': [29, 23, 37, 31],
      'city': ['Rome', 'Sydney', 'Tokyo', 'Moscow']}
df2 = pd.DataFrame(data2)

# 沿著列軸合併
print(pd.concat([df1, df2], axis=0))
```

```python
# 沿著行軸合併
print(pd.concat([df1, df2], axis=1))
```

####merge()

---
根據指定的欄位將兩個DataFrame物件合併在一起


```python
# 創建兩個DataFrame物件
data1 = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
      'age': [25, 32, 18, 47],
      'city': ['New York', 'Paris', 'London', 'Berlin']}
df1 = pd.DataFrame(data1)

data2 = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
     'salary': [5000, 6000, 7000, 8000]}
df2 = pd.DataFrame(data2)
```

```python
# 根據name欄位合併
print(pd.merge(df1, df2, on='name'))
```

####join()

---
將兩個DataFrame物件按照索引合併在一起


```python
# 創建兩個DataFrame物件
data1 = {'age': [25, 32, 18, 47],
      'city': ['New York', 'Paris', 'London', 'Berlin']}
df1 = pd.DataFrame(data1, index=['Alice', 'Bob', 'Charlie', 'David'])

data2 = {'salary': [5000, 6000, 7000, 8000]}
df2 = pd.DataFrame(data2, index=['Alice', 'Bob', 'Charlie', 'David'])
```

```python
# 按照索引合併
print(df1.join(df2))
```

####Groupby()

---

Groupby 是一個非常有用的功能，用於將 DataFrame 分組後對每個組進行操作

```python
data = {'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank'],
     'age': [25, 32, 18, 47, 23, 31],
     'gender': ['female', 'male', 'female', 'male', 'female', 'male'],
     'city': ['New York', 'Paris', 'London', 'Berlin', 'New York', 'Paris'],
     'salary': [50000, 80000, 25000, 100000, 35000, 75000]}
df = pd.DataFrame(data)
df
```

```python
# 按照 "city" 列的值進行分組，並對每個城市的人口數、平均年齡、平均薪資進行統計分析
grouped = df.groupby('city')
result = grouped.agg({'name': 'count', 'age': 'mean', 'salary': 'mean'})
print(result)
```

```python
grouped = df.groupby(['city', 'gender'])
result = grouped.agg({'name': 'count', 'age': 'mean', 'salary': 'mean'})
print(result)
```
