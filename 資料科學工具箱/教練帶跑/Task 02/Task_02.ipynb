#作業 01【實作題】資料選取

---


請從給定的 DataFrame 分別篩選出符合條件的資料：

1. filtered by first column > 20?

2. filtered by first column + second column > 50

3. filtered by first column < 30 or second column > 30

4. filtered by total sum of row > 100

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
```

##1.filtered by first column > 20?

```python
df[df.iloc[:,0]>20]

#or

df[df[0]>20]
```

##2.filtered by first column + second column > 50

```python
f2 = df.iloc[:,0] + df.iloc[:,1]
df[f2>50]

#or

f2 = df[0]+df[1]
df[f2>50]
```

##3.filtered by first column < 30 or second column > 30

```python
f3 = (df.iloc[:,0] < 30) | (df.iloc[:,1] > 30)
df[f3]
```

##4.filtered by total sum of row > 100

```python
f4 = df.sum(axis = 1) > 100
df[f4]
```

#作業 02【實作題】資料匯總

---

請利用 Pandas 計算給定的資料中不同 weekday 的 visitor 總和為何？

```python
d = [
    {'city': 'Austin', 'visitor': 139, 'weekday': 'Sun'},
    {'city': 'Dallas', 'visitor': 237, 'weekday': 'Sun'},
    {'city': 'Austin', 'visitor': 326, 'weekday': 'Mon'},
    {'city': 'Dallas', 'visitor': 456, 'weekday': 'Mon'}
]

df2 = pd.DataFrame(d)
df2
```

```python
gpweekday = df2.groupby('weekday')['visitor'].sum()
# gpweekday
res = gpweekday.to_dict()
print(res)
```

```python
grouped = df2.groupby('weekday')
result = grouped.agg({'visitor':'sum'})
result.to_dict()['visitor']
# result
```

```python
h1 = 'https://data.nhi.gov.tw/resource/Nhi_Fst/Fstdata.csv'
df3 = pd.read_csv(h1)
```

```python
res = df3.sort_values(by='快篩試劑截至目前結餘存貨數量', ascending=False).head()
res
# res 列出符合的每個colums
res.loc[:,['醫事機構地址','快篩試劑截至目前結餘存貨數量']]

# 縣市 vs count
res['縣市'] = res['醫事機構地址'].str[:3]
res.loc[:,['縣市','快篩試劑截至目前結餘存貨數量']]
```
