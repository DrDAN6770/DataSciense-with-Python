1. 練習：將 `winequality-both` 資料的後 10 筆資料當成「測試資料」、其餘資料作為「訓練資料」
2. 想要利用除了 type、quality 之外的欄位（X） 對 quality 欄位（y） 進行迴歸分析。請試著完成以下的程式碼
3. 並且解釋 A、B、C 三行程式碼分別在做什麼。

```python
import pandas as pd
import statsmodels.api as sm

source = 'https://raw.githubusercontent.com/cbrownley/foundations-for-analytics-with-python/master/statistics/winequality-both.csv'
df = pd.read_csv(source)
df
```

```python
# 取後10
test = df.tail(10)
# 因變數y
testy = test['quality']
# 自變數x, 將type、quality以外的欄位都作為自變數，並加上一個常數項
testX = sm.add_constant(test[test.columns.difference(['type','quality'])])

# 取後10以外
train = df.iloc[:-10,:] # or head(-10)
y = train['quality']
X = sm.add_constant(train[train.columns.difference(['type','quality'])])

model = sm.OLS(y,X).fit() # A : 建立迴歸模型

print(model.summary()) # B : 摘要
print(model.predict(testX)) # C : 用測試資料來預測模型與實際y值的差異
```

```python
# !pip install scikit-learn

from sklearn.metrics import mean_absolute_error, mean_squared_error

# 將預測結果和實際 y 值合併成一個 DataFrame
df1 = pd.concat([testy,model.predict(testX)], axis=1)
df1.columns = ['quality', 'y_pred']

# 計算 MSE 和 MAE
mse = mean_squared_error(df1['quality'], df1['y_pred'])
mae = mean_absolute_error(df1['quality'], df1['y_pred'])

# 將 MSE 和 MAE 轉換為準確度
accuracy_mse = 100 - mse
accuracy_mae = 100 - mae

print(f"Accuracy (MSE): {accuracy_mse:.2f}%")
print(f"Accuracy (MAE): {accuracy_mae:.2f}%")
```
