#作業02 資料清理處理策略

---

請利用題目提供的資料，分別完成以下幾種策略的資料填補方法：
* Drop missing observations
* Drop rows where all cells in that row is NA
* Create a new column full of missing values
* Fill in missing data with zeros
* Fill in missing in preTestScore with the mean value of preTestScore
* Fill in missing in postTestScore with each sex’s mean value of postTestScore
* Select some rows but ignore the missing data points

```python
import numpy as np
import pandas as pd
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'], 
        'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'], 
        'age': [42, np.nan, 36, 24, 73], 
        'sex': ['m', np.nan, 'f', 'm', 'f'], 
        'preTestScore': [4, np.nan, np.nan, 2, 3],
        'postTestScore': [25, np.nan, np.nan, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'sex', 'preTestScore', 'postTestScore'])
df
```

```python
# Drop missing observations
df.dropna()
```

```python
# Drop rows where all cells in that row is NA
df.dropna(how='all')
```

```python
# Create a new column full of missing values
dfmissingvalues = df.copy()
dfmissingvalues['Missing Values'] = np.nan
dfmissingvalues
```

```python
# Fill in missing data with zeros
df.fillna(0)
```

```python
# Fill in missing in preTestScore with the mean value of preTestScore
df_preTestScore_fill = df.copy()
df_preTestScore_fill['preTestScore'].fillna(df['preTestScore'].mean(),inplace = True)
df_preTestScore_fill
```

```python
# Fill in missing in postTestScore with each sex’s mean value of postTestScore
df_postTestScore_fill = df.copy()
df_postTestScore_fill['postTestScore'] = df.groupby('sex')['postTestScore'].transform(lambda x:x.fillna(x.mean()))
df_postTestScore_fill
```

```python
# Select some rows but ignore the missing data points
mask = df['first_name'].notnull()
df[mask]
```

#作業03 資料串接與資料清理

---
How to check if a dataframe has any missing values？Check if df has any missing values
* Count the number of missing values in each column and find the maximum * number of missing values.
* How to replace missing values of `multiple numeric` columns with the mean?
* Calculate the average price of different manufacturers.
* How to replace missing values of `price` columns with the mean depending on its manufacturers?


```python
import pandas as pd
url = 'https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'
df = pd.read_csv(url)
```

```python
df
```

```python
df.shape
```

```python
# Count the number of missing values in each column and find the maximum * number of missing values.
def missingvalues_count_max(df):
  return df.isnull().sum(),df.isnull().sum().max()

res = missingvalues_count_max(df)
print(f'count by col:\n{res[0]}\n\nmax:{res[1]}')
```

```python
# How to replace missing values of multiple numeric columns with the mean?
dfmissvalues = df.copy()
dfmissvalues.fillna(df.mean(),inplace = True)
dfmissvalues

#檢查
print(dfmissvalues.isnull().sum().sum())
print(df.isnull().sum().sum())
```

```python
# Calculate the average price of different manufacturers.
df_bymanufacturers = df.groupby('Manufacturer')['Price'].mean()
df_bymanufacturers
```

```python
# How to replace missing values of price columns with the mean depending on its manufacturers?
# index 78、85 Price missing

df_missingPrice = df.copy()
df_missingPrice['Price'] = df_missingPrice['Price'].fillna(df_missingPrice['Manufacturer'].map(df_bymanufacturers))

#檢查
df_missingPrice.iloc[[78,85]]
# df[df['Price'].isnull()]
```

```python
url = 'https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'
df = pd.read_csv(url)
# Create a copy of the original DataFrame
df_missingPrice = df.copy()

# Calculate the mean price for each manufacturer
manufacturer_mean = df.groupby('Manufacturer')['Price'].transform('mean')

# Replace the missing 'Price' values with the mean price for the respective manufacturer
df_missingPrice['Price'].fillna(manufacturer_mean, inplace=True)
df_missingPrice.iloc[[78,85]]
```
